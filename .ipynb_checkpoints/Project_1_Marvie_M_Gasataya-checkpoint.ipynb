```python
# Handling the Outliers

# We will apply data transformation to cap/floor the outliers using IQR

# Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'sepal width'
Q1 = df['sepal width (cm)'].quantile(0.25)
Q3 = df['sepal width (cm)'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Cap the outliers
df['sepal width (cm)'] = df['sepal width (cm)'].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

# Step 4: Display the updated dataset and boxplot
print("Updated dataset with outliers capped:")
df.boxplot(column=['sepal width (cm)'])
plt.show()
```

### a. Formulate one or two hypotheses based on the EDA findings.
    - Petal Length and Petal Width are strong predictors for species classification.
    - Sepal Width is not a significant distinguishing feature between the Iris species.
    - Iris Virginica flowers tend to have the largest overall flower dimensions (sepal length, petal length, and petal width).

### b. Use SciPy (or anything else you wish) to conduct appropriate statistical analysis. You could explore SciPy t-tests, chi-square tests, etc.

We have 5 identified hypothesis 3 from our EDA and 2 hypothesis just by obesreving the differences in data from the tables. Let us identify the Null hypotheses and the Alternate Hypothesis for each. Some of the hypothesis formed in our initial observation overlap with the hypothesis formed with our EDA. We have combined similar hypothesis and noted 5 distinct ones.

1. *Petal Length and Petal Width are strong predictors for species classification.*
    - Null Hypothesis (H₀): There is no significant difference in the mean petal length and petal width across the three species.
    - Alternative Hypothesis (H₁): There is a significant difference in the mean petal length and petal width across the three species.
    - This hypothesis can be tested using ANOVA or t-tests to see if petal length and width significantly differ between species, as suggested by the bimodal distributions and clear separation in the scatter plots.
     
2. *Sepal Width is not a significant distinguishing feature between the Iris species.*
   - Null Hypothesis (H₀): The mean sepal width is the same across all three species (Setosa, Versicolor, Virginica).
   - Alternative Hypothesis (H₁): The mean sepal width differs across at least one pair of species. 
   - This can be tested using ANOVA to determine whether there is a statistically significant difference in sepal widths between the three species.
     
3. *Iris Virginica flowers tend to have the largest overall flower dimensions (sepal length, petal length, and petal width).*
   - Null Hypothesis (H₀): There is no relationship between the size (sepal and petal dimensions) and the likelihood of being Iris Virginica.
   - Alternative Hypothesis (H₁): Larger flowers (longer sepal and petal lengths/widths) are more likely to be Iris Virginica.
   - This hypothesis can be tested by comparing the overall size of Iris Virginica flowers to the other species using t-tests.
   
4. *Iris Setosa flowers have smaller petal widths compared to the other two species.*
   - Null Hypothesis (H₀): The mean petal width of Iris Setosa is equal to the mean petal width of Iris Versicolor and Iris Virginica combined.
   - Alternative Hypothesis (H₁): The mean petal width of Iris Setosa is smaller than the mean petal width of Iris Versicolor and Iris Virginica combined.
   - This hypothesis can be tested with a t-test, comparing the mean petal width of Iris Setosa against the combined petal widths of the other two species.
     
5. *A smaller flower might be an indicator of Iris Setosa.*
   - Null Hypothesis (H₀): There is no relationship between the size (sepal and petal dimensions) and the species of the flower.
   - Smaller flowers (shorter sepal and petal lengths/widths) are more likely to be Iris Setosa.
   - We can test this hypothesis by comparing the sepal and petal sizes of Iris Setosa with the other species using t-tests or ANOVA.

```python
# Let us start with the EDA findings and calculate the statistical data
```

#### 1. Petal Length and Petal Width are strong predictors for species classification.
    - Null Hypothesis (H₀): There is no significant difference in the mean petal length and petal width across the three species.
    - Alternative Hypothesis (H₁): There is a significant difference in the mean petal length and petal width across the three species.

```python
from scipy import stats
import pandas as pd

# Combine Versicolor and Virginica for the comparison with Setosa
non_setosa = pd.concat([versicolor, virginica])

# Perform t-test for Petal Length (Setosa vs. Non-Setosa)
t_stat_petal_length, p_val_petal_length = stats.ttest_ind(setosa['petal length (cm)'], non_setosa['petal length (cm)'], equal_var=False)

# Perform t-test for Petal Width (Setosa vs. Non-Setosa)
t_stat_petal_width, p_val_petal_width = stats.ttest_ind(setosa['petal width (cm)'], non_setosa['petal width (cm)'], equal_var=False)

# Display the results
print("T-test for Petal Length (Setosa vs. Non-Setosa):")
print(f"T-statistic: {t_stat_petal_length:.6f}, P-value: {p_val_petal_length:.10f}")

print("\nT-test for Petal Width (Setosa vs. Non-Setosa):")
print(f"T-statistic: {t_stat_petal_width:.6f}, P-value: {p_val_petal_width:.10f}")
```

#### 2. Sepal Width is not a significant distinguishing feature between the Iris species.
   - Null Hypothesis (H₀): The mean sepal width is the same across all three species (Setosa, Versicolor, Virginica).
   - Alternative Hypothesis (H₁): The mean sepal width differs across at least one pair of species. 

```python
from scipy import stats

# Separate the species
setosa = df[df['species'] == 0]
versicolor = df[df['species'] == 1]
virginica = df[df['species'] == 2]

# Perform one-way ANOVA on Sepal Width across the three species
f_stat_sepal_width, p_val_sepal_width = stats.f_oneway(setosa['sepal width (cm)'], versicolor['sepal width (cm)'], virginica['sepal width (cm)'])

# Display the results
print("ANOVA for Sepal Width (Across Species):")
print(f"F-statistic: {f_stat_sepal_width:.6f}, P-value: {p_val_sepal_width:.10f}")
```

#### 3. Iris Virginica flowers tend to have the largest overall flower dimensions (sepal length, petal length, and petal width).
   - Null Hypothesis (H₀): There is no relationship between the size (sepal and petal dimensions) and the likelihood of being Iris Virginica.
   - Alternative Hypothesis (H₁): Larger flowers (longer sepal and petal lengths/widths) are more likely to be Iris Virginica.

```python
# Combine Setosa and Versicolor for comparison against Virginica
non_virginica = pd.concat([setosa, versicolor])

# Perform t-tests for Sepal Length, Petal Length, and Petal Width
t_stat_sepal_length, p_val_sepal_length = stats.ttest_ind(virginica['sepal length (cm)'], non_virginica['sepal length (cm)'], equal_var=False)
t_stat_petal_length, p_val_petal_length = stats.ttest_ind(virginica['petal length (cm)'], non_virginica['petal length (cm)'], equal_var=False)
t_stat_petal_width, p_val_petal_width = stats.ttest_ind(virginica['petal width (cm)'], non_virginica['petal width (cm)'], equal_var=False)

# Display the results
print("T-test for Sepal Length (Virginica vs. Non-Virginica):")
print(f"T-statistic: {t_stat_sepal_length:.6f}, P-value: {p_val_sepal_length:.10f}")

print("\nT-test for Petal Length (Virginica vs. Non-Virginica):")
print(f"T-statistic: {t_stat_petal_length:.6f}, P-value: {p_val_petal_length:.10f}")

print("\nT-test for Petal Width (Virginica vs. Non-Virginica):")
print(f"T-statistic: {t_stat_petal_width:.6f}, P-value: {p_val_petal_width:.10f}")
```

#### 4. Iris Setosa flowers have smaller petal widths compared to the other two species.
   - Null Hypothesis (H₀): The mean petal width of Iris Setosa is equal to the mean petal width of Iris Versicolor and Iris Virginica combined.
   - Alternative Hypothesis (H₁): The mean petal width of Iris Setosa is smaller than the mean petal width of Iris Versicolor and Iris Virginica combined.

```python
# Combine Versicolor and Virginica for comparison
non_setosa = pd.concat([versicolor, virginica])

# Perform one-sided t-test for Petal Width (Setosa vs Non-Setosa)
# The 'alternative' argument is used to specify a one-sided test.
t_stat_petal_width, p_val_petal_width = stats.ttest_ind(setosa['petal width (cm)'], non_setosa['petal width (cm)'], equal_var=False)

# Since we want to test if Setosa's petal width is smaller, we halve the p-value for the one-sided test
one_sided_p_val = p_val_petal_width / 2 if t_stat_petal_width < 0 else 1 - (p_val_petal_width / 2)

# Display the results
print("One-sided T-test for Petal Width (Setosa vs. Non-Setosa):")
print(f"T-statistic: {t_stat_petal_width:.6f}, One-sided P-value: {one_sided_p_val:.10f}")
```

#### 5. A smaller flower might be an indicator of Iris Setosa.
   - Null Hypothesis (H₀): There is no relationship between the size (sepal and petal dimensions) and the species of the flower.
   - Smaller flowers (shorter sepal and petal lengths/widths) are more likely to be Iris Setosa.

```python
# Combine Versicolor and Virginica for comparison
non_setosa = pd.concat([versicolor, virginica])

# Perform t-tests for Sepal Length, Sepal Width, Petal Length, and Petal Width
t_stat_sepal_length, p_val_sepal_length = stats.ttest_ind(setosa['sepal length (cm)'], non_setosa['sepal length (cm)'], equal_var=False)
t_stat_sepal_width, p_val_sepal_width = stats.ttest_ind(setosa['sepal width (cm)'], non_setosa['sepal width (cm)'], equal_var=False)
t_stat_petal_length, p_val_petal_length = stats.ttest_ind(setosa['petal length (cm)'], non_setosa['petal length (cm)'], equal_var=False)
t_stat_petal_width, p_val_petal_width = stats.ttest_ind(setosa['petal width (cm)'], non_setosa['petal width (cm)'], equal_var=False)

# Display the results
print("T-tests for Sepal and Petal Dimensions (Setosa vs. Non-Setosa):")

print(f"Sepal Length T-statistic: {t_stat_sepal_length:.6f}, P-value: {p_val_sepal_length:.10f}")
print(f"Sepal Width T-statistic: {t_stat_sepal_width:.6f}, P-value: {p_val_sepal_width:.10f}")
print(f"Petal Length T-statistic: {t_stat_petal_length:.6f}, P-value: {p_val_petal_length:.10f}")
print(f"Petal Width T-statistic: {t_stat_petal_width:.6f}, P-value: {p_val_petal_width:.10f}")
```

### c. Interpret the results of the statistical tests to draw conclusions about the data.

### For each test:
- If the p-value < 0.05, we reject the null hypothesis, indicating a significant difference between groups.
- If the p-value > 0.05, we fail to reject the null hypothesis, suggesting no significant difference.
